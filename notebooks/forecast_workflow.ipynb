# Forecast workflow notebook

This notebook demonstrates running the forecasting pipeline using the code in `src/forecast_pipeline.py`.
Make sure to create a `.env` file (see `.env.example`) and provide a SQL query file before running.

```python
from src.forecast_pipeline import DataLoader, VolumeForecastPipeline, ForecastVisualizer, build_conn_str_from_env
import os

# Ensure SQL_PATH env var is set or adjust the path below
sql_path = os.getenv('SQL_PATH', 'query.sql')
conn = build_conn_str_from_env()

loader = DataLoader(conn, sql_path)
df = loader.load()

pipeline = VolumeForecastPipeline(horizon=13)
forecast_df, weights = pipeline.run(df)

print(forecast_df.round(2))
print(weights)

# Uncomment to show interactive plot (requires a display or Jupyter environment)
# ForecastVisualizer.plot(df, forecast_df)
```
